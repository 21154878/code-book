```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()
```

```python
import operator
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures

# transforming the data to include another axis
x = X[:, np.newaxis]
y = Y[:, np.newaxis]
```

```python
polynomial_features= PolynomialFeatures(degree=2) # A fit of polynomial degree 2
x_poly = polynomial_features.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)

rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
r2 = r2_score(y,y_poly_pred)
print(rmse)
print(r2)

```

```python
plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

```python
# transforming the data to include another axis
x8 = X[:, np.newaxis]
y8 = Y[:, np.newaxis]

polynomial_features= PolynomialFeatures(degree=8) # A fit of polynomial degree 8
x_poly = polynomial_features.fit_transform(x8)

model = LinearRegression()
model.fit(x_poly, y8)
y_poly_pred = model.predict(x_poly)

rmse = np.sqrt(mean_squared_error(y8,y_poly_pred))
r2 = r2_score(y8,y_poly_pred)
print(rmse)
print(r2)

plt.scatter(x8, y8, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x8,y_poly_pred), key=sort_axis)
x8, y_poly_pred = zip(*sorted_zip)
plt.plot(x8, y_poly_pred, color='m')
plt.show()
```
