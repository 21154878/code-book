After clicking on the red point by the left button, you can get the photo of the building, and by using the right button to click on, you can get the name of the building.

```python
import pygame, sys
import random


# Define Game parameters
# screen size
screen_width = 1793/1.5
screen_height = 1486/1.5
FONT_NAME = pygame.font.match_font('arial')

# initialise the game window
pygame.init()
pygame.display.set_caption('La Defense')

# set the game surface
screen = pygame.display.set_mode((screen_width, screen_height))

# set the size of the unit
unit_width=screen_width/11
unit_height=screen_height/9

# a clock to keep track of the game progress
clock = pygame.time.Clock()

# Use existing images
backgrImage = pygame.image.load('plan of la defense.jpg')
backgrImage = pygame.transform.scale(backgrImage, (screen_width, screen_height))

screen.blit(backgrImage, [0,0])





# Update before the first frame
pygame.display.update()

def text(surf, text, size, x, y):
    font = pygame.font.Font(FONT_NAME, size)
    text_surface = font.render(text, True, 'black')
    text_rect = text_surface.get_rect()
    text_rect.midtop = (x, y)
    surf.blit(text_surface, text_rect)

# The unit rectangles
rectangles = []
for i in range(11):
    row = []
    for j in range(9):
        r = pygame.Rect(i * unit_width, j * unit_height,
                        unit_width, unit_height)
        row.append(r)
    rectangles.append(row)


# The game loop, in here the behaviour of the game is defined.
# This loop is executed every frame.
while True:
    
    events = pygame.event.get()
    
    for event in events:
        # Check exit through x button window
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

        elif event.type == pygame.KEYDOWN:
            # Check exit through esc
            if event.key == pygame.K_ESCAPE:
                pygame.quit()
                sys.exit()


        # The unit is filled with grey when moving the mouse
        elif event.type == pygame.MOUSEMOTION:
            (x,y)=event.pos
            screen.blit(backgrImage, [0,0])
            for i in range(12):
                for j in range(10):
                    if x > i*unit_width and x < (i+1)*unit_width and y > j*unit_height and y < (j+1)*unit_height:
                        pygame.draw.rect(screen, pygame.Color(90,90,90), rectangles[i][j])
            
        elif event.type == pygame.MOUSEBUTTONDOWN:
            (x,y)=event.pos
            # Clicking by the left button will get the photo.
            if event.button == 1:
                if x > 3*unit_width and x < 4*unit_width and y > 3*unit_width and y < 4*unit_height:
                    unit_image = pygame.image.load('unit/' + '3-3.png')
                    pygame.draw.rect(screen, pygame.Color(100,100,100), rectangles[3][3])
                    screen.blit(unit_image, [x,y])
                if x > 2*unit_width and x < 3*unit_width and y > 4*unit_width and y < 5*unit_height:
                    unit_image = pygame.image.load('unit/' + '2-4.png')
                    pygame.draw.rect(screen, pygame.Color(100,100,100), rectangles[2][4])
                    screen.blit(unit_image, [x,y])
                if x > 3*unit_width and x < 4*unit_width and y > 5*unit_width and y < 6*unit_height:
                    unit_image = pygame.image.load('unit/' + '3-5.png')
                    pygame.draw.rect(screen, pygame.Color(100,100,100), rectangles[3][5])
                    screen.blit(unit_image, [x,y])
                if x > 5*unit_width and x < 6*unit_width and y > 3*unit_width and y < 4*unit_height:
                    unit_image = pygame.image.load('unit/' + '5-3.png')
                    pygame.draw.rect(screen, pygame.Color(100,100,100), rectangles[5][3])
                    screen.blit(unit_image, [x,y])
                if x > 7*unit_width and x < 8*unit_width and y > 3*unit_width and y < 4*unit_height:
                    unit_image = pygame.image.load('unit/' + '7-3.png')
                    pygame.draw.rect(screen, pygame.Color(100,100,100), rectangles[7][3])
                    screen.blit(unit_image, [x,y])
            
            # Clicking by the right button will get the name of the building.
            if event.button == 3:
                if x > 3*unit_width and x < 4*unit_width and y > 3*unit_width and y < 4*unit_height:
                    screen.blit(backgrImage, [0,0])
                    text(screen,"CNIT",40,x,y)
                if x > 2*unit_width and x < 3*unit_width and y > 4*unit_width and y < 5*unit_height:
                    screen.blit(backgrImage, [0,0])
                    text(screen,"La Grande Arche",40,x,y)
                if x > 3*unit_width and x < 4*unit_width and y > 5*unit_width and y < 6*unit_height:
                    screen.blit(backgrImage, [0,0])
                    text(screen,"Les Quatre Temps",40,x,y)
                if x > 5*unit_width and x < 6*unit_width and y > 3*unit_width and y < 4*unit_height:
                    screen.blit(backgrImage, [0,0])
                    text(screen,"Coeur Defense",40,x,y)
                if x > 7*unit_width and x < 8*unit_width and y > 3*unit_width and y < 4*unit_height:
                    screen.blit(backgrImage, [0,0])
                    text(screen,"D2",40,x,y)


    pygame.display.update()
    clock.tick(60)
    
```
